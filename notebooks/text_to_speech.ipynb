# Text To Speech  

In this notebook, the Gemini Text to Speech API is tested, the main objective is to generate an audio file based on a transcript, which then will be added to a video

Most of the code was adapted from [here](https://ai.google.dev/gemini-api/docs/speech-generation?hl=es-419&authuser=1)

To use the Gemini API, you need to create an API-KEY, you can check the documentation [here](https://ai.google.dev/gemini-api/docs/api-key?hl=es-419#windows)

```python
from google import genai
from google.genai import types
import wave
```

```python
# Set up the wave file to save the output:
def wave_file(filename, pcm, channels=1, rate=24000, sample_width=2):
   with wave.open(filename, "wb") as wf:
      wf.setnchannels(channels)
      wf.setsampwidth(sample_width)
      wf.setframerate(rate)
      wf.writeframes(pcm)
```
